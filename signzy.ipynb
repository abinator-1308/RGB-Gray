```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import savefig
import cv2
```

```python
num_images = 3670
dataset = []

for i in range(1, num_images+1):
    img = cv2.imread("../input/color-images/color_images/color_images/color_" +str(i) +".jpg" )
    #print(np.array(img).shape)
    dataset.append(np.array(img))

dataset_source = np.asarray(dataset)
print(dataset_source.shape)

dataset_tar = []

for i in range(1, num_images+1):
    img = cv2.imread("../input/gray-images/gray_images/gray_images/gray_" +str(i) +".jpg", 0)    
    dataset_tar.append(np.array(img))

dataset_target = np.asarray(dataset_tar)
#To make the dimension of the gray images equal to the color images
dataset_target = dataset_target[:, :, :, np.newaxis]
print(dataset_target.shape)
```

```python
red = dataset_source[:,:,:,0]
green = dataset_source[:,:,:,1]
blue = dataset_source[:,:,:,2]
gray = dataset_target[:,:,:,0]
print(gray.shape)
```

```python
red = np.array(red.flatten())
green = np.array(green.flatten())
blue = np.array(blue.flatten())
gray = np.array(gray.flatten())
print(red.shape)
print(gray)
```

```python
y = gray
X = np.column_stack((red,green,blue))
print(X.shape)

#initialize weight with random values
w = np.array([0.3,0.8,0.2])
print(w.shape)
Cost = []
Weights = []
alpha = 0.001
for i in range(0,100):
    pred =  X[:,0]*w[0] + X[:,1]*w[1] + X[:,2]*w[2]
    pred = pred.astype(int)
    err = np.array(y-pred)
    sq_err = np.array(np.square(err))
    print("Cost = ", np.median(sq_err)/256)
    Cost.append(np.median(sq_err)/256)
    print("Weights = ", w)
    Weights.append(w)
    print(i)
    w[0] = w[0] - alpha*(np.median(sq_err)/256)
    w[1] = w[1] - alpha*(np.median(sq_err)/256)
    w[2] = w[2] - alpha*(np.median(sq_err)/256)
    
```

```python
#predict
img = cv2.imread("../input/color-images/color_images/color_images/color_19.jpg")
print(np.array(img).shape)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img_nn = img[:,:,0]*w[0]+img[:,:,1]*w[1]+img[:,:,2]*w[2]
img_nn = img_nn.astype(int)
print(img_nn.shape)
plt.imshow(img_nn)
plt.show()
```
